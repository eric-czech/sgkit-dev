```python
def read(path, batch_size, n=None):
    ct = 0
    with open_bgen(path) as bgen:
        if n is None:
            n = bgen.nvariants
        print(f'Found {n} variants')
        for i in range(0, n, batch_size):
            # Shape returned is (n_samples, n_variants, n_genotypes)
            val = bgen.read(slice(i, i + batch_size))
            ct += val.shape[0] * val.shape[1]
    print(f'Number of entries read: {ct}')
```

```python
with ResourceProfiler() as prof:
    read(path, 1000, n=10000)
```
